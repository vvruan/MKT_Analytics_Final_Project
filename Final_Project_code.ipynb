# A/B test

```python
# to calculate conversion rate, we use proportion
import numpy as np
from scipy.stats import norm

def calc_proportion(variant):
    return sum(variant)/len(variant)

# calculate z-score
def calc_zscore(p_hat, p, n_f):
    return (p_hat-p)/np.sqrt(p*(1-p)/n_f)

# calculate probability 
def get_z_crit_value(alpha_f, num_sides_f):
    return norm.ppf(1-(alpha_f/num_sides_f))

# get p value for z-alpha
def get_p_value(zscore_f, num_sides_f):
    return (1-norm.cdf(abs(zscore_f))) * num_sides_f

# build hypothesis test and analyze if we can reject null 
def hypothesis_test(variantA_outcomes_f, variantB_outcomes_f, alpha_f, num_sides_f):
    p_hat_f = calc_proportion(variantB_outcomes_f)
    p_f = calc_proportion(variantA_outcomes_f)
    n_f = len(variantB_outcomes_f)
    z_score = calc_zscore(p_hat_f, p_f, n_f)
    p_value = get_p_value(z_score, num_sides_f)
    z_crit = get_z_crit_value(alpha_f, num_sides_f)
    return z_score > z_crit, z_score, p_value
```

```python
# set assumptions
alpha = 0.05 # set confidence level alpha
num_sides = 1
confidence_level = 1 -(alpha/num_sides)
```

```python
# only look at top three plan type because of their data volumes
conversion_by_plan_type = pd.DataFrame(index = ["conversion_rate","reject","z_score","p_value"], \
                        columns=['base_uae_14_day_trial','high_uae_14_day_trial','low_uae_no_trial'])
```

```python
plan_type_list = ['base_uae_14_day_trial','high_uae_14_day_trial','low_uae_no_trial']

for i in range(len(plan_type_list)):
    variantA_base_uae_14 = subscribers.loc[(subscribers['plan_type'] == 'base_uae_14_day_trial'),'conversion']
    variantB = subscribers.loc[(subscribers['plan_type'] == plan_type_list[i]),'conversion']
    reject_null_test, z_score, p_value = hypothesis_test(variantA_base_uae_14, variantB, alpha, num_sides)
    
    conversion_by_plan_type.at["conversion_rate",plan_type_list[i]] = round(calc_proportion(variantB),4)
    conversion_by_plan_type.at["reject",plan_type_list[i]] = reject_null_test
    conversion_by_plan_type.at["z_score",plan_type_list[i]] = z_score
    conversion_by_plan_type.at["p_value",plan_type_list[i]] = p_value
```

# Churn

```python
subscribers_copy = subscribers.copy()

#drop uses who fill in age > 70, taken as random fillins
subscribers_copy.drop(subscribers_copy[subscribers_copy["age"]>70].index,inplace=True)

#drop timestamp columns
subscribers_copy.drop(["account_creation_date","trial_end_date"],axis = 1,inplace=True)

#drop rows where join_fee < 0, because they doesn't look like real customers
subscribers_copy.drop(subscribers_copy.loc[subscribers_copy['join_fee'] < 0].index, inplace=True)

#drop corelated columns
subscribers_copy.drop(['refund_after_trial_TF','paid_TF','cancel_before_trial_end',\
                       'creation_until_cancel_days','payment_type',"revenue_net",\
                      "initial_credit_card_declined"],axis = 1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

df_train,df_test = train_test_split(subscribers_copy,test_size = 0.2, random_state = 11)

y_train = (df_train.conversion).values
y_test = (df_test.conversion).values
```

```python
del df_train['conversion']
del df_test['conversion']
```

```python
# one-hot encoding
from sklearn.feature_extraction import DictVectorizer
df_train = df_train.fillna(0)
df_test = df_test.fillna(0)

dict_train = df_train.to_dict(orient='records')
dict_val = df_test.to_dict(orient='records')

dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(dict_train)
X_val = dv.transform(dict_val)
```

```python
# use decision tree model to make prediction
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
# model tuning 
for m in [8,10,12]:
    print('depth: %s' % m)

    for s in [1, 5, 10, 15, 20, 50, 100, 200]:
        dt = DecisionTreeClassifier(max_depth=m, min_samples_leaf=s)
        dt.fit(X_train, y_train)
        y_pred = dt.predict_proba(X_val)[:, 1]
        auc = roc_auc_score(y_test, y_pred)
        print('%s -> %.3f' % (s, auc))
    
    print()
```

```python
# Build discount strategy model 
def apply_offers(df, threshold, base_price, discount_amount, cases, acceptance_rate_churn, acceptance_rate_nonchurn):
    df_f = df.copy()

    # set offer parameters for all observations
    df_f['price_no_offer'] = base_price
    df_f.loc[df_f['conversion'] == False, 'price_no_offer'] = 0  # if the person churns they will pay 0

    # determine whether indiviudals accept offer
    df_f.loc[df_f['conversion'] == False, 'accept_offer_prob'] = acceptance_rate_churn  # those who were going to churn are less likely to want the offer than those who were going to stay
    df_f.loc[df_f['conversion'] == True, 'accept_offer_prob'] = acceptance_rate_nonchurn
    df_f['would_accept_offer_rand_draw'] = (df_f['random_draws'] < df_f['accept_offer_prob'])  # randomly select the individuals who would accept an offer based on their probability of acceptance

    # make offer under lpm predictions for select threshold
    df_f['make_offer'] = (df_f['prob_churn'] > threshold)     # predict whether individual will churn or not
    df_f['accept_offer'] = (df_f['make_offer'] & df_f['would_accept_offer_rand_draw'])
    df_f['price_offer'] = df_f['price_no_offer']  # initialize prices from "do nothing case"
    df_f.loc[df_f['accept_offer'], 'price_offer'] = base_price * (1-discount_amount)    # for those who were both made an offer and would accept, set prices to discounted price

    df_impact = pd.DataFrame(index=cases, columns=['subs', 'revenue', 'lift_from_no_offer'])
    for c in cases:
        df_impact.loc[c, 'subs'] = (df_f['price_' + c] > 0).sum() # 
        df_impact.loc[c, 'revenue'] = df_f['price_' + c].sum()
    for c in cases:
        df_impact.loc[c, 'lift_from_no_offer'] = df_impact.loc[c, 'revenue'] / df_impact.loc['no_offer', 'revenue'] - 1

    return df_f, df_impact
```
